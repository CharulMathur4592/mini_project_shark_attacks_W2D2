```python
# droping Unnamed 21 & Unnamed 22 columns

new_df = df.drop(['pdf', 'href formula', 'href', 'Unnamed: 21', 'Unnamed: 22'], axis = 1)
new_df
```

```python
new_df.columns = (
    new_df.columns
        .str.strip()              # remove leading/trailing spaces
        .str.replace(" ", "_")  # replace internal spaces with _
)
new_df.columns
```

```python
# To clean the column "Sex"

def clean_sex_col(x):

    s = str(x).strip().lower()

    if s.startswith("m"):
        return "M"
    if s.startswith("f"):
        return "F"
    else:
        return "others"
    
new_df["Sex"] = new_df["Sex"].apply(clean_sex_col)
```

```python

# Convert everything to string for cleaning
new_df["Age"] = new_df["Age"].astype(str).str.lower().str.strip()

# Replace common text ages with approximate numeric values
age_replacements = {
    "teen": 15,
    "teenager": 15,
    "adult": 30,
    "young adult": 25,
    "mid-age": 40,
    "middle aged": 45,
    "middle-aged": 45,
    "mid 30s": 35,
    "30s": 35,
    "40s": 45,
    "50s": 55,
    "60s": 65,
    "elderly": 70,
    "senior": 70
}


new_df["Age"] = new_df["Age"].replace(age_replacements)

# Handle "unknown", "nan", empty strings, and invalid symbols
new_df["Age"] = new_df["Age"].replace(["unknown", "?", "nan", "", "none"], "unknown")

# Convert ranges like "20-30" → take the midpoint (25)
new_df["Age"] = new_df["Age"].apply(
    lambda x: (int(x.split("-")[0]) + int(x.split("-")[1])) / 2 
    if isinstance(x, str) and "-" in x and x.replace("-", "").isdigit()
    else x
)

new_df["Age"] = new_df["Age"].apply(
    lambda x: int(x.replace("s", "")) + 5 
    if isinstance(x, str) and x.endswith("s") and x[:-1].isdigit()
    else x
)

new_df["Age"] = pd.to_numeric(new_df["Age"], errors="coerce")

# Optional: remove unrealistic ages
new_df.loc[(new_df["Age"] < 0) | (new_df["Age"] > 100), "Age"] = np.nan

print(new_df["Age"].describe())
```

```python
import re

# Step 1: Convert to string, lowercase, remove leading/trailing spaces
new_df["Name"] = new_df["Name"].astype(str).str.lower().str.strip()

# Step 2: Replace known invalid names with NaN
invalid_terms = [
    "nan", "unknown", "no name", "n/a", "none", "unidentified", "?", "",
    "male", "female", "boy", "girl", "man", "woman", "child", "person"
]
new_df["Name"] = new_df["Name"].replace(invalid_terms, "Unknown")

# Step 3: Remove text in parentheses (e.g. “(surname unknown)”)
new_df["Name"] = new_df["Name"].apply(lambda x: re.sub(r"\(.*?\)", "", x) if isinstance(x, str) else x)

# Step 4: Remove punctuation
new_df["Name"] = new_df["Name"].str.replace(r"[^a-zA-Z\s']", "", regex=True)

# Step 5: Remove multiple spaces
new_df["Name"] = new_df["Name"].str.replace(r"\s+", " ", regex=True).str.strip()

# Step 6: Remove names that are still too short or meaningless
new_df.loc[new_df["Name"].apply(lambda x: len(x) < 3 if isinstance(x, str) else False), "Name"] = np.nan

# Step 7: Capitalize name properly (title case)
new_df["Name"] = new_df["Name"].str.title()

# Step 8: Optional — keep only the first name if more text remains
# Example: "John Smith Jr" → "John Smith"
new_df["Name"] = new_df["Name"].apply(lambda x: " ".join(x.split()[:2]) if isinstance(x, str) else x)
```
