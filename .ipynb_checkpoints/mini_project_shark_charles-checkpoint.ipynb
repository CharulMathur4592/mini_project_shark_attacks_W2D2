```python
new_df = df.drop(['pdf', 'href formula', 'href', 'Unnamed: 21', 'Unnamed: 22'], axis = 1)
new_df
```

```python
# Ensure Species is string
species =new_df["Species "].dropna().astype(str)

# Basic cleaning (remove extra spaces)
species = species.str.strip().str.replace(r"\s+", " ", regex=True)

# Keep only entries that look like real shark types:
# Rule: must contain the word 'shark' and be reasonably short
valid_sharks = species[
    (species.str.contains("shark", case=False)) &
    (species.str.len() < 50)   # removes long descriptive sentences
]

new_df["clean species"] = valid_sharks
```

```python
new_df["clean species"] = new_df["clean species"].fillna("Unknown")
```

```python
df.columns = [ col.strip().replace(" ", "_").lower() for col in list(df.columns) ]
df.columns
```

```python
new_df["clean species"] # List of valid shark species

valid_sharks = [
    "White shark",
    "Tiger shark",
    "Bull shark",
    "Blacktip shark",
    "Nurse shark",
    "Wobbegong shark"
]

# Ensure Species column is clean (remove spaces and normalize)
df["species"] = df["species"].astype(str).str.strip()

# Categorise species
df["species_grouped"] = df["species"].apply(
    lambda x: x if x in valid_sharks else (x if x == "Unknown" else "Other")
)

# Optional: view counts
print(df["species_grouped"].value_counts())
```
