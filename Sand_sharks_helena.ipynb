# Type

```python
new_df["Type"].value_counts()
```

```python
new_df["Type_clean"] = new_df["Type"].str.strip().str.lower()
```

```python
new_df["Type_clean"].value_counts()
```

```python
def simple_type(s):
    
    if s == "unprovoked":
        return "unprovoked"
    elif s == "provoked":
        return "provoked"
    else:
        return "other"
        
```

```python
new_df["Type_simple"] = new_df["Type_clean"].apply(simple_type)
```

```python
new_df["Type_simple"].value_counts()
```

# Activity

```python
#hehe
new_df["Activity"].value_counts().tail(10)
```

```python
new_df["Activity_clean"] = new_df["Activity"].str.strip().str.lower()
```

```python
new_df["Activity_clean"] = new_df["Activity_clean"].fillna("unknown")
```

```python

def simple_activity(s):
    
    if "swim" in s:             
        return "swimming"
    elif "surf" in s:
        return "surfing"
    elif "fish" in s:
        return "fishing"
    elif "wading" in s:
        return "wading"
    elif "diving" in s:
        return "diving"
    elif "snorkel" in s:
        return "snorkeling"
    elif "kayak" in s:
        return "kayaking"
    elif "bath" in s:
        return "bathing"
    elif "sail" in s:
        return "sailing"
    elif "board" in s:
        return "boarding"
    elif "paddl" in s:
        return "paddling"
    elif "unknown" in s:
        return "unknown"
    else:
        return "other"
```

```python
new_df["Activity_simple"] = new_df["Activity_clean"].apply(simple_activity)
new_df["Activity_simple"].value_counts()
```

```python
new_df[new_df["Activity_simple"]=="other"][["Activity","Activity_simple"]].head(10)
```

```python
swimming_related = ["swimming", "bathing", "wading"]
```

```python
water_sports = [
    "surfing",
    "diving",
    "snorkeling",
    "paddling",
    "kayaking",
    "paddling",
    "boarding",
    "fishing",
    "sailing"
]
```

```python
unknown = ["unknown"]
```

```python
final_activity = []

for x in new_df["Activity_simple"]:
    if x in swimming_related:
        final_activity.append("swimming_related")
    elif x in water_sports:
        final_activity.append("water_sports")
    elif x in unknown:
        final_activity.append("unknown")
    else:
        final_activity.append("other")

new_df["Activity_final"] = final_activity
```

```python
new_df["Activity_final"].value_counts()
```

# Fatal

```python
new_df["Fatal_Y/N"].value_counts()
```

```python
new_df["Fatal_clean"] = new_df["Fatal_Y/N"].str.strip().str.lower()
```

```python
new_df["Fatal_clean"] = new_df["Fatal_clean"].fillna("unknown")
```

```python
new_df["Fatal_clean"].value_counts()
```

```python
def simple_fatal(x):
    s = str(x).strip().lower()
    if s == "y":
        return "fatal"
    else:
        return "non_fatal"

```

```python
new_df["Fatal_simple"] = new_df["Fatal_clean"].apply(simple_fatal)
new_df["Fatal_simple"].value_counts()
```

## Fatal by type

```python
# count only fatal by type
fatal_by_type = new_df[new_df["Fatal_simple"] == "fatal"]["Type_simple"].value_counts()
fatal_by_type
```

```python
total_by_type = new_df["Type_simple"].value_counts()
total_by_type
```

```python
# Percentage of dead by type
percent_fatal_by_type = fatal_by_type / total_by_type * 100
percent_fatal_by_type
```

```python
summary = pd.DataFrame({"Total": total_by_type,"Fatal": fatal_by_type,"Percent_fatal": percent_fatal_by_type})
summary
```

```python
summary = summary.sort_values("Percent_fatal", ascending=False)
summary["Percent_fatal"] = summary["Percent_fatal"].round(2)
summary
```

```python
summary_activity = pd.DataFrame({
    "Total": new_df["Activity_simple"].value_counts(),
    "Fatal": new_df[new_df["Fatal_simple"] == "fatal"]["Activity_simple"].value_counts()
})

summary_activity["Percent_fatal"] = (summary_activity["Fatal"] / summary_activity["Total"] * 100).round(2)
summary_activity.sort_values("Percent_fatal", ascending=False)
```

```python
summary_activity_new = pd.DataFrame({
    "Total": new_df["Activity_simple"].value_counts(),
    "Fatal": new_df[new_df["Fatal_simple"] == "fatal"]["Activity_simple"].value_counts()
})
```

```python
summary_activity_new["Fatal"] = summary_activity_new["Fatal"].fillna(0)

summary_activity_new["Percent_fatal"] = (
    summary_activity_new["Fatal"] / summary_activity_new["Total"] * 100).round(2)

summary_activity_new
```

```python
import matplotlib.pyplot as plt

summary_activity_new["Percent_fatal"].plot(kind="bar")

plt.title("Fatality rate by activity")
plt.ylabel("Percentage fatality")
plt.xlabel("Activity")
plt.xticks(rotation=45)

plt.show()
```

```python
# Ordenar de més a menys perillós
summary_activity_new = summary_activity_new.sort_values("Percent_fatal", ascending=False)
summary_activity_new["Percent_fatal"].plot(kind="bar")

# Títols
plt.title("Fatality rate by activity")
plt.ylabel("Percent fatal")
plt.xlabel("Activity")
plt.xticks(rotation=45, ha="right")

plt.show()
```
